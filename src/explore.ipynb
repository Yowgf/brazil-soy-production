```python
df_productivity = lib.process.prod(df)
df_productivity = df_productivity[df_productivity.region == 'Brasil']
lib.visualize(df_productivity)
```

### Productivity from **1960** to **2006**

```python
prod_1960_2006 = df_productivity[
    (df_productivity["year"] >= 1960)
    & (df_productivity["year"] <= 2006)
][["year", "productivity"]]
prod_1960_2006["year"] = prod_1960_2006["year"] - 1960
prod_1960_2006
```

### Productivity from **2006** onwards

```python
prod_2006_2023 = df_productivity[
    (df_productivity["year"] >= 2006)
    & (df_productivity["year"] <= 2023)
][["year", "productivity"]]
prod_2006_2023["year"] = prod_2006_2023["year"] - 2006
prod_2006_2023
```
